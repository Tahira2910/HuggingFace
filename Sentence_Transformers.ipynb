**Generate Embedding**

```python
# used for transforming sentences into fixed-dimensional vector representations.
# offers a range of functionalities to support various natural language processing (NLP) tasks that involve
# sentence embeddings, such as semantic similarity, clustering, information retrieval, and more based on their embeddings.

from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
model
```

```python
sentences = ['This framework generates embeddings for each input sentence',
            'Sentences are passed as a list of string.']

embeddings = model.encode(sentences)

print(embeddings)
```

```python
# iterate through embeddings on our sentences
for sentence, embedding in zip(sentences, embeddings):
  print("Sentence: ", sentence)
  print("Embedding: ", embedding)
  print("")
```

**Cosine Similarity**

```python
# Cosine Similarity-->  Operation to check similarity between vectors
```

```python
emb1 = model.encode("I am eating apple")
emb2 = model.encode("I like fruits")
# utility module has this cosine_similarity function which takes
# embeddings of two sentences and it tells you what is the cosine-sim
cos_sim = util.cos_sim(emb1, emb2)
print("Cosine-Similarity:", cos_sim)
```

**Compute Cosine Similarity between all pairs**

```python
# Compute cosine similarity between all pairs

sentences = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'The girl is carrying a baby.',
          'A man is riding a horse.',
          'A woman is playing violin.',
          'Two men pushed carts through the woods.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.'
          ]
# Encode all sentences
embeddings = model.encode(sentences)

# Compute cosine similarity between all pairs
# how each of the sentence embedding is similar to each of the other.
cos_sim = util.cos_sim(embeddings, embeddings)

cos_sim
```

```python
# to find which sentence is more similar to other sentence
# Add all pairs to a list with their cosie similarity score
all_sentence_combinations = []
for i in range(len(cos_sim)-1):
  for j in range(i+1, len(cos_sim)):
    all_sentence_combinations.append((cos_sim[i][j], i, j))
all_sentence_combinations
```

**Clustering**

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
# embedder--> model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Corpus with example sentences
corpus = ['A man is eating food.',
          'A man is eating a piece of bread.',
          'Horse is eating grass.',
          'A man is eating pasta.',
          'A Woman is eating Biryani.',
          'The girl is carrying a baby.',
          'The baby is carried by the woman',
          'A man is riding a horse.',
          'A man is riding a white horse on an enclosed ground.',
          'A monkey is playing drums.',
          'Someone in a gorilla costume is playing a set of drums.',
          'A cheetah is running behind its prey.',
          'A cheetah chases prey on across a field.',
          'The cheetah is chasing a man who is riding the horse.',
          'man and women with their baby are watching cheetah in zoo'
          ]

corpus_embeddings = embedder.encode(corpus)

# Normalize the embeddings to unit length
# By normalization becomes easy for distance based matrix or optimization kind of thing
# we are going to calculate the distance between the two vectors(any 2 sentences in corpus)
# Doesn't matters much if don't normalize
corpus_embeddings = corpus_embeddings / np.linalg.norm(corpus_embeddings, axis = 1, keepdims=True)
```

```python
# source: https://stackoverflow.com/questions/55619176/how-to-cluster-similar-sentences-using-bert

clustering_model = KMeans(n_clusters=3)
# data--> embeddings created
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python
# creating a dictionary and put all the sentences for that particular cluster into the one list

clustered_sentences = {}
# cluster_assignment--> list of clusters
# cluster_id--> one particular cluster
# sentence_id--> number(index) associated with it
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])
clustered_sentences
```
